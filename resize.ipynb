```
 def resize(images_hr):
            lr_size=(28,28)
            channels=3
            images_lr = np.empty(shape=(images_hr.shape[0],) + lr_size + (channels,))
            for i in range(images_hr.shape[0]):
                images_lr[i] = cv2.resize(images_hr[i], dsize=(28, 28), interpolation=cv2.INTER_CUBIC)

            return images_lr
```

```
resized=resize(imgs)
```
